```python
import pandas as pd
```

```python
df = pd.read_csv("data/Bellevue_Permits.csv")
```

```python
df
```

```python
df = df[['X', 'Y', 'SITEADDRESS', 'CITY', 'ZIPCODE',
         'PERMITSTATUS', 'PERMITYEAR','PROJECTNAME', 'PROJECTDESCRIPTION', 'ISSUEDDATE', 'TOTALPAID', 
         'SUBAREA', 'NEIGHBORHOODAREA', 'SUBTYPE', 'ZONING', 'LOTSIZE']]
```

```python
df['ZONING'].value_counts()
```

```python
filter2 = df['ZONING'].isin(["[O] Office", '[OLB] Office and Limited Business', "[BR-OR-1] Bel-Red-Office/Residential Node 1", "[BR-OR-2] Bel-Red-Office/Residential Node 2", '[BR-MO] Bel-Red-Medical Office', '[BR-OR] Bel-Red Office/Residential', '[BR-ORT] Bel-Red-Office/Residential Transition', '[PO] Professional Office', '[BR-MO-1] Bel-Red-Medical Office Node','[OLB-OS] Office & Limited Business-Open Space' ])
```

```python
df = df[filter2]
```

```python
df = df[df.PERMITSTATUS != 'Expired']
df = df[df.PERMITSTATUS != 'On Hold']
df = df[df.PERMITSTATUS != 'Screening']
df = df[df.PERMITSTATUS != 'Monitoring']
df = df[df.PERMITSTATUS != 'Incomplete']
df = df[df.PERMITSTATUS != 'Screened Out']
df = df[df.PERMITSTATUS != 'Canceled']
df = df[df.PERMITSTATUS != 'TCO']
```

```python
df = df.loc[df['PERMITYEAR'] == 2021]
```

```python
df = df.loc[df['SUBTYPE'] == 'Nonresidential']
```

```python
df.rename(columns={"X": "Longitude", "Y": "Latitude"})
```

```python
df.to_csv(r'data\dataset.csv', index = False)
```
